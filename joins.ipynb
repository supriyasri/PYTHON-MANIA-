```python
import pandas as pd 
table_1={'Region':['BI','CAFRICA','CA','EAFRICA','EA','EASTERNEUROPE','ME','NA'],
        'Code':['GBR','COD','MEX','ETH','JPN','RUS','TUR','USA'],
        'Name':['UK','Congo','Mexico','Ethiopia','Japan','Russia','Turkey','United states'],
        'Population':['59','51','98','62','126','146','66','278']}
df1=pd.DataFrame(table_1)
```

```python
print(df1)
df1
```

```python
table_2={'Region':['BI','CA','EARFICA','EA','EA','EE','EE','ME'],
        'Code':['GBR','MEX','ETH','CHN','JPN','RUS','UKR','TUR'],
        'Surface area':['242','1958','110','957','377','1707','603','774'],
        'Head of state':['Eizabeth','Vicento','Negasso','Jiang','Akhio','Putin','Leoni','Ahmet']}
df2=pd.DataFrame(table_2)
```

```python
df2
```

```python
pd.merge(df1,df2)               
# Region and Code are two common columns in both the tables. This is similar to intersection of two tables.
# Inner join displayed
```

```python
pd.merge(df1,df2,how='inner')


```

```python
pd.merge(df1,df2,how='outer')   #all entires do come ,equivalent to union 
```

```python
pd.merge(df1,df2,how='left')   #all rows in left table (df1) are present, id there is missing value in right table
# it appears as NA 
```

```python
pd.merge(df1,df2,how='right')
```

```python
df1.rename(columns={'Code':'CD'},inplace=True)  #give the exact column name to change 
print(df1, '\n',df2)
```

```python
pd.merge(left=df1,right=df2,left_on='CD',right_on='Code',how='inner')
```

```python
pd.merge(left=df1,right=df2,left_on='CD',right_on='Code',how='outer')
```

```python
pd.merge(left=df1,right=df2,left_on='CD',right_on='Code',how='left')  
```

```python
pd.merge(left=df1,right=df2,left_on='CD',right_on='Code',how='left',suffixes=('_left','_right')) 
#suffix for similar column name is clearly displayes as Region_left, Region_right
```

```python
df1.set_index('CD',inplace=True)
df2.set_index('Code',inplace=True)   #we have taken set_index for two dataframes
display(df1,df2)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
#validate merge

#used to check whether duplication is present or not

df2.reset_index(inplace=True)
df2
```

```python
dummy=df2[0:2]
x=pd.concat([dummy]*2)  #creates two rows of dummy variables to the existing dummy variable
x
```

```python
display(df1,x)    #here X do contain duplicate code-GBR 
```

```python
pd.merge(df1,x,left_on='CD',right_on='Code',how='inner',validate='one_to_many')

#left dataframe (df1) only one GBR but right dataframe(x) do have two GBR. so one_to_many holds.
```
